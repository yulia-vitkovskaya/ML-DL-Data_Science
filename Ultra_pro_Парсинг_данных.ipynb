```
df = pd.read_csv('/content/drive/My Drive/БД/moscow.csv', sep = ';')
```

```
df.head(20)
```

```
df_ = df[::2]
df_ = df_.reset_index(drop = True)
df_ = df_.drop(['ГРМ', 'Источник', 'Примечание'], axis = 1)
df_.index = pd.to_datetime(df_['Дата'])
df_.drop('Дата', axis=1, inplace=True)
tripTime = df_['От станции']
tripTime = tripTime.fillna('Нет данных')
for i in range(tripTime.shape[0]):
  if 'т' in tripTime[i]:
    tripTime[i] = tripTime[i].replace('т', ' минут на транспорте')
  elif 'п' in tripTime[i]:
    tripTime[i] = tripTime[i].replace('п', ' минут пешком')
df_['От станции'] = tripTime
cost = df_['Цена, руб.']
for i in range(cost.shape[0]):
  cost[i] = cost[i] / 1000000
df_['Цена, руб.'] = cost
```

```
df_ = df_.rename(columns = {'Цена, руб.':'Цена, млн. руб.'})
```

```
arr = []
house = df_['Дом']
for i in range(house.shape[0]):
  ind = house[i].find('/')
  el = house[i]
  arr.append(el[:ind])
  if 'П' in house[i]:
    house[i] = 'Панельный'
  elif 'М' in house[i]:
    house[i] = 'Монолитный'
  elif 'К' in house[i]:
    house[i] = 'Кирпичный'
  else:
    house[i] = 'Нет данных'
df_['Этаж'] = arr
```
