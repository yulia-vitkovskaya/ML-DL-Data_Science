### Задача 2

Перед вами стоит задача посчитать финансовые расходы за месяц. Ежедневно вы фиксировали свои траты и в итоге составили список расходов:

        purchases = [1200, 800, 468, 235, 5800, 1350, 110, 243,
                    767, 3500, 5400, 4389, 3690, 2420, 894,
                    1766, 2100, 450, 328, 1890, 233, 766, 1765,
                    237, 679, 1983, 389, 1760, 2100, 253, 789]
Чтобы не ошибиться и точно посчитать свои траты, произведите вычисления тремя разными способами:
1. С помощью цикла `while`;
2. С помощью цикла `for`;
3. С помощью встроенной функции `sum()`.

---
Ваш результат должен быть таким:

img21.jpg

```python
# Список расходов
purchases = [1200, 800, 468, 235, 5800, 1350, 110, 243,
            767, 3500, 5400, 4389, 3690, 2420, 894,
            1766, 2100, 450, 328, 1890, 233, 766, 1765,
            237, 679, 1983, 389, 1760, 2100, 253, 789]
```

```python
# Ваше решение
purchases = [1200, 800, 468, 235, 5800, 1350, 110, 243,
            767, 3500, 5400, 4389, 3690, 2420, 894,
            1766, 2100, 450, 328, 1890, 233, 766, 1765,
            237, 679, 1983, 389, 1760, 2100, 253, 789]

amount = 0
for num in purchases:
  amount += num
print(f'Траты за месяц составили: {amount} рубля')

amount = 0
i = 0
while i < len(purchases):
  amount += purchases[i]
  i+=1
print(f'Траты за месяц составили: {amount} рубля')

amount = sum(purchases)
print(f'Траты за месяц составили: {amount} рубля')
```

### Задача 3

Вы уже знаете, что кортежи являются неизменяемыми, но как быть, если нам нужно удалить элемент кортежа? Создайте функцию `del_from_tuple(tpl, elem)`, которая будет удалять первое появление определенного элемента (**elem**) из кортежа (**tpl**) и возвращать кортеж без него.

Если в кортеже (**tpl**) нет указанного элемента (**elem**), то функция должна вернуть кортеж в исходном виде.

---

В зависимости от введенных данных ваш результат может быть таким:

pic4.jpg

```python
# Ваше решение
def del_from_tuple(tpl, elem):
  if elem in tpl:
    n = tpl.index(elem)
    new_tuple = tpl[n+1:] if (n == 0) else tpl[:n] + tpl[n+1:]
  else:
    new_tuple = tpl
  return new_tuple

#Вариант 1
tpl = (1, 2, 3)
elem = 1
print(f'Кортеж: {tpl}. Элемент: {elem}')
print(f'Кортеж без заданного элемента: {del_from_tuple(tpl, elem)}')

#Вариант 2
tpl = (1, 2, 3, 1, 2, 3, 4, 5, 2, 3, 4, 2, 4, 2)
elem = 3
print(f'Кортеж: {tpl}. Элемент: {elem}')
print(f'Кортеж без заданного элемента: {del_from_tuple(tpl, elem)}')

#Вариант 3
tpl = (2, 4, 6, 6, 4, 2)
elem = 9
print(f'Кортеж: {tpl}. Элемент: {elem}')
print(f'Кортеж без заданного элемента: {del_from_tuple(tpl, elem)}')
```

### Задача 4

Дана текстовая фраза:

        phrase = "В чащах юга жил бы цитрус? Да, но фальшивый экземпляр!"

Ваша задача: посчитать и вывевести на экран количество символов (с учетом пробелов и знаков препинания) до буквы **е**.

Подсказка: к символам строковой переменной можно обращаться так же, как к элементам списка.

---

Ваш результат должен быть таким:
img24.jpg

```python
# присваивание переменной phrase значения
phrase = "В чащах юга жил бы цитрус? Да, но фальшивый экземпляр!"
```

```python
# Ваше решение
num = 0
phrase = "В чащах юга жил бы цитрус? Да, но фальшивый экземпляр!"
for letter in phrase:
  if letter != 'е':
    num +=1
  else:
    break
print(f'Количество символов во фразе до буквы "е": {num}')
```

### Задача 5

Создайте функцию `to_list()`, которая принимает неограниченное количество параметров.
Обработайте их так, чтобы на выходе получить список из этих элементов.

В зависисмости от введенных данных ваш результат может выглядеть так:

---

image.png

```python
# Ваше решение
def to_list(*args):
  return list(args)

tpl_1 = (1, 2, 3)
print(f'Кортеж аргументов: {tpl_1}')
print(f'Список элементов: {to_list(tpl_1)}\n')

tpl_2 = ('Молоко', 5, '2020 год')
print(f'Кортеж аргументов: {tpl_2}')
print(f'Список элементов: {to_list(tpl_2)}\n')

tpl_3 = ([3, 4, 7], 8.3, True, 'Строка')
print(f'Кортеж аргументов: {tpl_3}')
print(f'Список элементов: {to_list(tpl_3)}\n')
```
