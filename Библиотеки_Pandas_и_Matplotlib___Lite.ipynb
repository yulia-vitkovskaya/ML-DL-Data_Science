Ниже дано 2 графика. Сделайте их более простыми для восприятия:
- подпишите оси - "ось X" и "X и Y"
- подпишите название графика - "sin(x)/2 и cos(x)"
- увеличьте размер полотна

---

Ваш результат должен быть таким:

Answer_1.jpg

````python
# Исходные графики:

import matplotlib.pyplot as plt    # Импорт модуля pyplot
import numpy as np                 # Импорт библиотеки numpy

x = np.linspace(0, 10, 101)        # Генерация последовательности из 101 числа с шагом в 0.1
y = np.cos(x)                      # Запись cos(x) в переменную y
z = np.sin(x) / 2                  # Запись sin(x) / 2 в переменную z

plt.plot(x, y)                     # Построение графика x, y
plt.plot(x, z)                     # Построение графика x, z
plt.show()                         # Фиксация графика
````

````python
# Ваше решение
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 101)
y = np.cos(x)
z = np.sin(x) / 2

plt.figure(figsize=(10, 3))
plt.plot(x, y)
plt.plot(x, z)
plt.xlabel('Ось X')
plt.ylabel('X и Y')
plt.title('sin(x)/2 и cos(x)')
plt.show()
````

### Задача 5

Добавьте к датафрейму новый столбец, который будет хранить в себе данные по средней температуре в градусах по Цельсию (сейчас столбец **AvgTemperature** хранит температуру в градусах по Фаренгейту). Эта задача решается без использования циклов.

Формула для перевода градусов Фаренгейта в градусы Цельсия:
```
(°F − 32) × 5/9 = °C
```

---

Ваш результат должен быть таким:

````python
new_df['AvgTemperature_C'] = ((new_df.AvgTemperature - 32) * 5/9)
````
