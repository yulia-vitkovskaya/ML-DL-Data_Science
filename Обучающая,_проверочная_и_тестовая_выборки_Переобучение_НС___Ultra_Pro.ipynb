```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
import keras.utils as image
import matplotlib.pyplot as plt
from tensorflow.keras import utils
import gdown, os, random
```

```python
# Ваше решение
import gdown

gdown.download('https://storage.yandexcloud.net/aiueducation/Content/base/l4/bus.zip', None, quiet = True)
```

```python
from zipfile import ZipFile
zipfile = 'bus.zip'
z = ZipFile(zipfile)
z.infolist()
```

```python
x_data = []
y_data = []
for files in os.listdir('/content/Входящий'):
  im = np.array(image.load_img('/content/Входящий/' + files, target_size=(128, 64)))
  x_data.append(im)
  y_data.append(0)
for files in os.listdir('/content/Выходящий'):
  im = np.array(image.load_img('/content/Выходящий/' + files, target_size=(128, 64)))
  x_data.append(im)
  y_data.append(1)
```

```python
x_data = np.array(x_data)
y_data = np.array(y_data)
print(x_data.shape)
print(y_data.shape)
```

```python
x_data = x_data.reshape(9081, 24576)
print(x_data.shape)
x_data = x_data.astype('float32')
x_data = x_data / 255
y_data = utils.to_categorical(y_data, 2)
print(y_data.shape)
print(y_data[12])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data,
                                                    y_data,
                                                    test_size=0.2,
                                                    shuffle=True,
                                                    random_state=3)
print('Обучающая выборка параметров', x_train.shape)
print('Обучающая выборка меток классов', y_train.shape)
print()
print('Тестовая выборка параметров', x_test.shape)
print('Тестовая выборка меток классов', y_test.shape)
```

```python
def model():
  model = Sequential()

  model.add(Dense(1500, input_dim = 24576, activation = 'relu'))
  model.add(Dense(500, activation = 'relu'))
  model.add(Dense(2, activation = 'sigmoid'))

  model.compile(loss="binary_crossentropy",
                optimizer=Adam(learning_rate=0.0001),
                metrics=["accuracy"])

  model.fit(x_train,
            y_train,
            batch_size=128,
            epochs=20,
            verbose=1,
            validation_split = 0.2,
            shuffle=True)

  scores = model.evaluate(x_test, y_test, verbose=1)
  print('Доля верных ответов на тестовых данных, в процентах: {:7.4%}'.format(scores[1]))
```

```python
model = model()
```
