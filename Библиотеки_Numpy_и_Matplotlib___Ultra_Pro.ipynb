### Задача 1

Напишите функцию матричного умножения трёх матриц без использования встроенных средств `np.dot()`, **@**.

```python
# Ваше решение
def multi(m1, m2, m3):
  shape1 = m1.shape
  shape2 = m2.shape
  shape3 = m3.shape

  if shape1[1] != shape2[0] or shape1[1] != shape3[0]:
    return np.array([])

  interim = np.zeros((shape1[1], shape2[0]))

  for i in range(shape1[0]):
   for j in range(shape2[1]):
     curr_cell = 0
     for t in range(shape1[1]):
       curr_cell += m1[i,t] * m2[t,j]
     interim[i,j] = curr_cell

  shape4 = interim.shape
  out = np.zeros((shape4[1], shape3[0]))

  for i in range(shape4[0]):
   for j in range(shape3[1]):
     curr_cell = 0
     for t in range(shape4[1]):
       curr_cell += interim[i,t] * m3[t,j]
     out[i,j] = curr_cell

  return out
```

```python
matrix1 = np.array([[4, 3], [7, 5]])
matrix2 = np.array([[-28, 93], [38, -126]])
matrix3 = np.array([[7, 3], [2, 1]])

print(multi(matrix1, matrix2, matrix3))
```

### Задача 2

Нужно вывести на печать матрицу размером N*N, заполненную числами натурального ряда, начиная с первого элемента и двигаясь по спирали:

---

В зависимости от введенных данных результат может выглядеть так:

image.png

```python
# Ваше решение

def fun(n):
  print('Размерность квадратной матрицы:', n)
  mat = np.zeros((n, n), dtype = int)
  st, m = 1, 0
  mat[n//2][n//2]=n*n
  for v in range(n//2):
    for i in range(n-m):
      mat[v][i+v] = st
      st+=1
    for i in range(v+1, n-v):
      mat[i][-v-1] = st
      st+=1
    for i in range(v+1, n-v):
      mat[-v-1][-i-1] =st
      st+=1
    for i in range(v+1, n-(v+1)):
      mat[-i-1][v]=st
      st+=1
    m+=2

  for i in mat:
    print(*i)
```

```python
fun(6)
```

```python
# Ваше решение
import random
import math
num = 100
x = [random.triangular() for i in range(num)]
y = [random.gauss(0.5, 0.25) for i in range(num)]
colors = [random.randint(1, 4) for i in range(num)]
areas = [math.pi * random.randint(5, 15)**2 for i in range(num)]
plt.figure()
plt.scatter(x, y, s=areas, c=colors, alpha=0.85)
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.show()
```
