```
def ParameterForSexAge(arg):
  if 'Мужчина' or 'Male' in arg:
    sex = 'М'
  elif 'Женщина' or 'Female' in arg:
    sex = 'Ж'
  else:
    sex = 'Не указано'

  cur_year = 2023
  year = 0
  if (len(arg) > 7):
    year = cur_year - int(re.findall(r'\d{4}', arg)[0])
  return sex, str(year)

def ParameterForSalary(arg):
  salString = 0
  salString = int(re.sub(r'[^0-9]*', '', arg))
  curr = re.sub(r'[0-9]*', '', arg)

  curr = curr.replace(' ', '').replace('.', '')
  if 'USD' in curr: salString = float(salString) * 84
  elif 'EUR' in curr: salString = float(salString) * 91
  elif 'грн' in curr: salString = float(salString) * 2,28
  elif 'белруб' in curr: salString = float(salString) * 28,13
  elif 'KZT' in curr: salString = float(salString) * 0,18
  elif 'KGS' in curr: salString = float(salString) * 0,96
  return salString

def ParameterForCity(arg):
  city = re.match(r'[^,]*', arg)
  city = city.group(0).replace(' ', '')

  removal = ''
  if ('не готов к переезду' in arg)| ('не готова к переезду' in arg): removal = 'Не готов(а)'
  elif ('готов к переезду' in arg)| ('готова к переезду' in arg): removal = 'Готов(a)'
  else: removal = 'Не указано'

  trip = ''
  if ('не готов к командировкам' in arg)| ('не готова к командировкам' in arg): trip = 'Не готов(а)'
  elif ('готов к командировкам' in arg)| ('готова к командировкам' in arg): trip = 'Готов(a)'
  else: trip = 'Не указано'

  return city, removal, trip

def ParameterForEmployment(arg):
  result = ''
  if ("стажировка" in arg)|("volunteer" in arg)|("work placement" in arg): result += 'стажировка '
  if ("частичная занятость" in arg)|("part" in arg): result += 'частичная занятость '
  if ("проектная работа" in arg)|("project" in arg): result += 'проектная работа '
  if ("полная занятость" in arg)|("full" in arg): result += 'полная занятость'
  if result=='':
    result='Не указано'
  return result

def ParameterForSchedule(arg):
  result = ''
  if ("гибкий график" in arg)|("flexible" in arg): result += 'гибкий график '
  if ("полный день" in arg)|("full" in arg): result += 'полный день '
  if ("сменный график" in arg)|("shift" in arg): result += 'сменный график '
  if ("удаленная работа" in arg)|("remote" in arg): result += 'удаленная работа'
  if result=='':
    result='Не указано'
  return result

def ParameterForExp(arg):
  sarg=arg.split(" ", 5)
  years = 0
  months = 0
  if len(sarg)!=6: sarg.append(0);sarg.append('');sarg.append(0);sarg.append('')
  try:
    years = int(sarg[2])
  except ValueError:
    years = 0

  try:
    months = int(sarg[4])
  except ValueError:
    months = 0
  return 12*years+months

def ParameterForEdu(arg):
  result = ''
  arg = arg.lower()
  if ("высшее" in arg)|("higher education"  in arg): result += 'высшее '
  if ("среднее специальное"  in arg) | ("special"  in arg) | ("колледж"  in arg) | ("college"  in arg) : result += 'среднее специальное '
  if ("неоконченное высшее"  in arg) | ("incomplete"  in arg): result += 'неоконченное высшее '
  if ("среднее образование"  in arg) | ("secondary education"  in arg): result += 'среднее образование'
  if result=='':
    result='Не указано'
  return result
```

```
hh = hh.drop(hh.columns[[2, 7, 8, 10, 11]], axis = 1)
hh.head(1)
```

```
funcList = [ParameterForSexAge, ParameterForSalary, ParameterForCity,
            ParameterForEmployment, ParameterForSchedule, ParameterForExp,
            ParameterForEdu]
for i in range(len(hh.columns)):
  hh[hh.columns[i]] =[ funcList[i](cell) for cell in hh[hh.columns[i]]]
```
