```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Дан следующий исходный график:

```python
x = np.linspace(0, 20, 101)               # Генерация последовательности из 101 числа с шагом в 0.2
y = np.cos(x)                             # Запись cos(x) в переменную y

plt.plot(x, y)                            # Построение графика x, y
plt.show()                                # Фиксация графика
```

Измените отображение графика таким образом, чтобы получился такой результат:

Graphic.jpg

- размер полотна: 12,4
- ширина линии: 5
- цвет линии: **4b1234**
- тип маркера: h
- размер маркера: 10
- цвет маркера: **ff00d5**
- периодичность маркера: 5
- легенда графика закреплена в правом нижнем углу: **cos(x)**
- заголовок графика: **cos(x)**
- подписи осей: **ось X** и **ось Y**
- отрисована сетка

```python
x = np.linspace(0, 20, 101)
y = np.cos(x)

plt.figure(figsize=(12, 4))
plt.plot(x, y, color = '#4b1234', linewidth = 5, marker = 'h', mfc = '#ff00d5', markersize = 10, markevery = 5, label = 'cos(x)')
plt.legend(loc = 'lower right')
plt.title('cos(x)')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid()
plt.show()
```
