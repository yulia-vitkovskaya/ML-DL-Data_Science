*2 балла*

Используя библиотеку Matplotlib, постройте графики функций $y = x^3$ и $y = x^2$ на одном графике.

* Дайте название графику
* Дайте названия осям
* Присвойте лейблы и выведите легенду
* Графики функций должны быть отрисованы пунктирными линиями (любыми из доступных на ваш вкус)

### **Решение:**

```
x = np.linspace(-3, 3, 50)
y1 = x**3
y2 = x**2
plt.plot(x, y1, '--', label='y = x^3')
plt.plot(x, y2, '.-', label='y = x^2')
plt.title('Графики x^2 и x^3')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.legend()
plt.show()
```

### **Задача 5**

*2 балла*

Постройте violinplot всех признаков датасета из прошлой задачи.

P.S. График должен быть читаемым. Для этого необходимо:
* отрегулировать размер графика,
* предварительно отскалировать все *числовые* признаки, т. е. привести данные к стандартному нормальному распределению при помощи питоновой функции scale, которая предоставляется ниже.

```
# Функция для скалирования
def scale(df):
    return (df - df.mean()) / df.std()
```

### **Решение:**

```
plt.figure(figsize = (10, 8))
sns.violinplot(data = scale(data), orient='h')
plt.show()
```
