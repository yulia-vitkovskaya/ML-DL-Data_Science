Задание 1

Модифицировать функцию Sine таким образом, чтобы можно было возвращать составной сигнал, состоящий из комбинаций sin и cos, каждая отдельная гармоническая функция в составном сигнале должна принимать свои параметры:

 A  = амплитуда сигнала

 f  = частота сигнала  Hz

 phi  = начальная фаза синусоиды "Радианы"

 Сделайте вариант с 2-мя и 3-мя гармоническими сигналами.

```
def Sine_2(**kwargs):
    fs = kwargs['fs']
    t = kwargs['t']
    N = kwargs['N']
    T = t/fs
    n = np.arange(N)

    x1 = kwargs['A1'] * np.sin(2 * np.pi * kwargs['f1'] * n * T + kwargs['phi1']) + kwargs['A2'] * np.cos(2 * np.pi * kwargs['f2'] * n * T + kwargs['phi2'])
    x2 =  kwargs['A1'] * np.sin(2 * np.pi * kwargs['f1'] * n * T + kwargs['phi1']) + kwargs['A2'] * np.cos(2 * np.pi * kwargs['f2'] * n * T + kwargs['phi2']) + kwargs['A3'] * np.sin(2 * np.pi * kwargs['f3'] * n * T + kwargs['phi3'])
    return x1, x2
```

```
x1, x2 = Sine_2(fs = 50, t = 5, N = 100, A1 = 1, A2 = 2, A3 = 3, f1 = 10, f2 = 20, f3 = 30, phi1 = 1, phi2 = 1, phi3 = 1)
```

```
plt.plot(x1)
plt.title('Sin + Cos')
plt.xlabel('Дискретные отсчёты')
plt.ylabel('Амплитуда')
plt.grid()
plt.show()
```

```
plt.plot(x2)
plt.title('Sin + Cos + Sin')
plt.xlabel('Дискретные отсчёты')
plt.ylabel('Амплитуда')
plt.grid()
plt.show()
```
